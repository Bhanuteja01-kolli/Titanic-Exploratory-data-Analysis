```python
# Outliers
for i in a:
  q1=a[i].quantile(0.25) # 7.9
  q2=a[i].quantile(0.50) # 14.4
  q3=a[i].quantile(0.75) # 31

iqr=q3-q1 # 31-8=23
uw=q3+1.5*iqr # 31+1.5*23=42
lw=q1-1.5*iqr # 31-11=20
print(q1)
print(q2)
print(q3)
```

```python
outliers=((a[i]>uw)|(a[i]<lw))
print(outliers.sum())
```

```python
a.loc[(a[i]>uw) | (a[i]<lw),i]=q2
print(outliers.sum())
```

# **Data Analysis**

```python
df.head()
```

```python
df['Survived'].value_counts()
```

most of the ppl survived in the **accident**

```python
x=df.groupby('Sex')['Survived'].mean()*100
# most of the female survived in the incident
x
```

```python
# binning
bins=[0,20,30,50,70]
labels=['child','young_adult','adult','old']
df['Agegroup']=pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
C=df.groupby('Agegroup')['Survived'].mean()*100
```

```python
# The childrens are survivins more in the incident
d=df.groupby('Pclass')['Survived'].mean()*100
```

```python
df.sort_values(by="Fare",ascending=False).head()
# who is paying are more surviving
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
plt.title("Titanic Data Analysis Dashboard")
plt.subplot(2,3,1)
sns.countplot(x="Survived",data=df,color="purple")
plt.subplot(2,3,2)
sns.countplot(x="Sex", hue="Survived", data=df)
plt.subplot(2,3,3)
plt.pie(C.values,labels=C.index,autopct="%1d%%")
plt.subplot(2,3,4)
plt.pie(d.values,labels=d.index,autopct="%1d%%")
plt.subplot(2,3,5)
sns.scatterplot(x="Age",y="Fare",hue="Survived",data=df)
plt.subplot(2,3,6)
sns.boxplot(x="Fare",data=df)
plt.show()
```
